```python
%%sql
-- 1. Write a query in SQL to find the name and year of the movies.¶
SELECT mov_title, mov_year
FROM movie;
```

```python
%%sql
-- 2. Write a query in SQL to find the year when the movie American Beauty released.
SELECT mov_year as year_released
FROM movie
WHERE mov_title='American Beauty';
```

```python
%%sql
-- 3. Write a query in SQL to find the movie which was released in the year 1999.
SELECT mov_title as movies_in_1999
FROM movie
WHERE mov_year=1999;
```

```python
%%sql
-- 4. Write a query in SQL to find the movies which was released before 1998.
SELECT mov_title as movies_before_1998
FROM movie
WHERE mov_year<1998;
```

```python
%%sql
-- 5. Write a query in SQL to return the name of all reviewers and name of movies together in a single list
-- Union
SELECT reviewer.rev_name
FROM reviewer
UNION
(SELECT movie.mov_title
FROM movie);
```

```python
%%sql
-- 6. Write a query in SQL to find the name of all reviewers who have rated 7 or more stars 
-- to their rating. | Comparing rating + reviewer table
SELECT reviewer.rev_name
FROM reviewer, rating
WHERE rating.rev_id = reviewer.rev_id
AND rating.rev_stars>=7 
AND reviewer.rev_name IS NOT NULL;
```

```python
%%sql
-- 7. Write a query in SQL to find the titles of all movies that have no ratings. 
-- Filtering elements present in one table but not in the second table
SELECT mov_title
FROM movie
WHERE mov_id NOT IN (
SELECT mov_id 
FROM rating
);
```

```python
%%sql
-- 8. Write a query in SQL to find the titles of the movies with ID 905, 907, 917. 
-- Filtering based on a given list of items
SELECT mov_title
FROM movie
WHERE mov_id in (905, 907, 917);
```

```python
%%sql
-- 9. Write a query in SQL to find the list of all those movies with year which 
-- include the words Boogie Nights. | String matching
SELECT mov_id, mov_title, mov_year
FROM movie
WHERE mov_title LIKE '%Boogie%Nights%'
ORDER BY mov_year ASC;
```

```python
%%sql
-- 10. Write a query in SQL to find the ID number for the actor whose first name 
-- is 'Woody' and the last name is 'Allen'. | Filtering using multiple conditions
SELECT act_id
FROM actor 
WHERE act_fname='Woody' 
AND act_lname='Allen';
```

```python
%%sql
-- 11. Write a query in SQL to find the name of all reviewers who have rated their ratings 
-- with a NULL value.
SELECT rev_name
FROM reviewer
INNER JOIN rating USING(rev_id)
WHERE rev_stars IS NULL;
```

```python
%%sql
-- 12. Write a query in SQL to list the first and last names of all the actors who were 
-- cast in the movie 'Annie Hall', and the roles they played in that production.
SELECT act_fname,act_lname,role
  FROM actor 
	  JOIN movie_cast ON actor.act_id=movie_cast.act_id
		JOIN movie ON movie_cast.mov_id=movie.mov_id 
		  AND movie.mov_title='Annie Hall';
```

```python
%%sql
-- 13. Write a query in SQL to find the name of movie and director (first and last names) 
-- who directed a movie that casted a role for 'Eyes Wide Shut'. | Join + Filter
SELECT dir_fname, dir_lname, mov_title
FROM  director 
NATURAL JOIN movie_direction
NATURAL JOIN movie
NATURAL JOIN movie_cast
WHERE role IS  NOT NULL
AND mov_title='Eyes Wide Shut';
```

```python
%%sql
-- 14. Write a query in SQL to find the name of movie and director (first and last names) 
-- who directed a movie that casted a role as Sean Maguire. | Join + Filter
SELECT dir_fname, dir_lname, mov_title
FROM  director 
JOIN movie_direction 
  ON director.dir_id=movie_direction.dir_id
JOIN movie 
  ON movie_direction.mov_id=movie.mov_id
JOIN movie_cast 
  ON movie_cast.mov_id=movie.mov_id
  WHERE role='Sean Maguire';
```

```python
%%sql
-- 15. Write a query in SQL to list all the actors who have not acted in any movie between 
-- 1990 and 2000. | Join + Filter
SELECT act_fname, act_lname, mov_title, mov_year
FROM actor
JOIN movie_cast 
ON actor.act_id=movie_cast.act_id
JOIN movie 
ON movie_cast.mov_id=movie.mov_id
WHERE mov_year NOT BETWEEN 1990 and 2000;
```
